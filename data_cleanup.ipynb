```python
import pandas as pd
df = pd.read_csv("job_offers_combined.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df["offer_name"] = df["offer_name"].apply(lambda x: x.replace("oferta pracy - ", "")).apply(lambda x: x.replace("job offer - ", ""))
```

```python
df
```

```python
import ast

df['salaries'] = df['salaries'].apply(ast.literal_eval)
df = df.explode("salaries")
df.rename(columns={"salaries": "salary"}, inplace=True)
```

```python
df
```

```python
df["contract_type"] = df.salary.str.extract(r'\((.*?)\)')
df["salary"] = df.salary.str.replace(r'\(.*?\)', '', regex=True)
```

```python
df.head()
```

```python
df["salary"] = df.salary.str.replace("zł", '')
df["salary"] = df.salary.str.replace(" ", "")
```

```python
df.head()
```

```python
df["salary_lower_end"] = df.salary.apply(lambda x: x.split("–")[0])
df["salary_upper_end"] = df.salary.apply(lambda x: x.split("–")[1])
```

```python
df["salary_lower_end"] = df["salary_lower_end"].str.replace(" ", "")
df["salary_upper_end"] = df["salary_upper_end"].str.replace(" ", "")
df["salary_lower_end"] = df["salary_lower_end"].apply(lambda x: float(x.replace("k", ""))*1000 if "k" in x else x)
df["salary_upper_end"] = df["salary_upper_end"].apply(lambda x: float(x.replace("k", ""))*1000 if "k" in x else x)
df["salary_lower_end"] = df["salary_lower_end"].astype(float)
df["salary_upper_end"] = df["salary_upper_end"].astype(float)
```

```python
df["salary_lower_end"] = df["salary_lower_end"].apply(lambda x: x*168 if x < 1000 else x)
df["salary_upper_end"] = df["salary_upper_end"].apply(lambda x: x*168 if x < 1000 else x)
```

```python
df.head()
```

```python
df["work_modes"].unique()
```

```python
mapping_dict = {
    'zdalna': ['remote'],
    'home office, hybrid': ["full_office", "hybrid"],
    'home office, hybrid, full office': ["remote", "hybrid", "full_office"],
    'home office': ['remote'],
    'hybrydowa': ['hybrid'],
    'hybrid': ['hybrid'],
    'stacjonarna': ['full_office'],
    'full office': ['full_office'],
    'hybrydowa, stacjonarna': ['hybrid', 'full_office'],
    'zdalna, hybrydowa': ['remote', 'hybrid'],
    'hybrid, full office': ['hybrid', "full_office"],
    'zdalna, hybrydowa, stacjonarna': ["remote",'hybrid', "full_office"],
    'zdalna, stacjonarna': ['remote', "full_office"],
    'mobilna': ['remote']
}

df["work_modes"] = df["work_modes"].map(mapping_dict)
```

```python
df["work_modes"].explode().unique()
```

```python
df.head()
```

```python
df["seniority"] = df["seniority"].apply(lambda x: [x.split(" • ")[0], x.split(" • ")[1]] if " • " in x else [x])
```

```python
df.head()
```

```python
df["office"] = df["office"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else [x])
df = df.explode("office")
df["office"] = df["office"].apply(lambda office: office.split(",")[0].strip())
```

```python
df.drop(columns=["salary", "whole_poland"], inplace=True)
```

```python
df["technologies"] = df["technologies"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else [x])
```

```python
df.head()
```

```python
df["contract_type"] = df["contract_type"].apply(lambda x: "UoP" if x == "CoE" else x)
df["contract_type"] = df["contract_type"].apply(lambda x: "zlec." if x == "CoM" else x)
```

```python
df["average_salary"] = (df["salary_lower_end"] + df["salary_upper_end"]) / 2
```
